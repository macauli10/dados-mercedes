```python
import pandas as pd
```

```python
df = pd.read_csv("../dados_silver_csv/v1_sem_duplicatas.csv")
df.head()
```

```python
gold_ml = df.copy()

gold_ml['horsepower_scaled'] = (gold_ml['horsepower'] - gold_ml['horsepower'].mean()) / gold_ml['horsepower'].std()
gold_ml['price_scaled'] = (gold_ml['base_price_usd'] - gold_ml['base_price_usd'].mean()) / gold_ml['base_price_usd'].std()
gold_ml['year_scaled'] = (gold_ml['year'] - gold_ml['year'].min()) / (gold_ml['year'].max() - gold_ml['year'].min())

gold_ml = pd.get_dummies(
    gold_ml, 
    columns=['model', 'color', 'fuel_type', 'region'], 
    prefix=['model', 'color', 'fuel', 'region'],
    drop_first=True  
)


features = gold_ml.drop(columns=['base_price_usd', 'turbo'])
target = gold_ml['base_price_usd']
features.to_parquet('../../gold/ml_features.parquet', compression='snappy')


target_df = pd.DataFrame(target)  
target_df.to_parquet('../../gold/ml_target.parquet', compression='snappy')

print(f"features ML: {features.shape[1]} colunas")
```
