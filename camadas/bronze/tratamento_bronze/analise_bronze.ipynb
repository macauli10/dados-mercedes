```python
df = pd.read_csv("dados_bruto.csv")
df.head(10)
```

```python
df.columns
```

```python
# arrumando os nomes das colunas
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
df = df.rename(columns={'base_price_(usd)': 'base_price_usd'})
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
total_registros = len(df)
registros_unicos = df.drop_duplicates().shape[0]
print(f"total: {total_registros}")
print(f"unicos: {registros_unicos}")
print(f"duplicatas: {total_registros - registros_unicos}")
```

```python
df.to_csv('../silver/v1_com_duplicatas.csv', index=False)
```

```python
df_sem_duplicata = df.drop_duplicates()
```

```python
df_sem_duplicata.duplicated().sum()
```

```python
df_sem_duplicata.to_csv('../silver/v1_sem_duplicatas.csv', index=False)
```

```python
df_sem_duplicata.to_parquet(
    '../silver/v1_sem_duplicatas.parquet', 
    index=False,
    compression='snappy'  
)
```
